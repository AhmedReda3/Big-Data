```python
import pandas as pd
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
total_cities_df =  pd.read_csv('/worldcities.csv', encoding='ISO-8859-1')
total_cities_df
```

```python
 Cities_df = total_cities_df[total_cities_df["country"].isin(['Italy', 'Egypt', 'United Kingdom','India','Spain','Germany','France','Canada','United States','Russia','Brazil'])]
                                   
 Cities_df
```

```python
Cities_df = Cities_df.dropna()
Cities_df
```

```python
Cities_df = Cities_df.drop(['lat', 'lng','iso2','admin_name','capital','population','country'], axis=1)
Cities_df = Cities_df.sort_values(by ='iso3')
Cities_df
```

```python
Cities_df.rename(columns={'city_ascii': 'city_name','iso3': 'country_code','id':'city_code'},inplace=True)
ran = np.random.uniform(low=19, high=25, size=len(Cities_df));
ran = ran.round(4)
Cities_df['average_temprature'] = ran
Cities_df
```

```python
countries_arr = pd.Series(['Italy', 'Egypt', 'United Kingdom','India','Spain','Germany','France','Canada','United States','Russia','Brazil'])
Code_arr = pd.Series(['ITA', 'EGY', 'GBR', 'IND','ESP', 'DEU', 'FRA', 'CAN', 'USA', 'RUS', 'BRA'])
Cdata = {
    'Country_Name':countries_arr,
    'Country_Code':Code_arr}
Country_df = pd.DataFrame(Cdata)
Country_df
```

```python
sensor_id = []
sensor_temp = []
country_code = []

for s in Cities_df['city_name'] :
  for i in range(1,6) :
    for i in range(1,6) :
      sensor_id.append(s +str(i))

for s in Cities_df['country_code'] :
  for i in range(1,6) :
    for i in range(1,6) :
      country_code.append(s)


for temp  in Cities_df['average_temprature']:
  for i in range(1,6) :
    for j in range(1, 6):
      old_value = temp 
      min_value = old_value - (0.01 * old_value)
      max_value = old_value + (0.01 * old_value)
      new_value = np.random.uniform(min_value, max_value)
      new_value = np.round(new_value, 4)
      time_data = '2222-05-10-0' + str(i) + ':00'
      sensor_temp.append((new_value,time_data))  

Sensor_data = pd.DataFrame(sensor_temp,columns = ['Temperture', 'Datetime'])
Sensor_id = pd.DataFrame(sensor_id,columns = ['Sensor_ID'])

readings_df = pd.concat([Sensor_id, Sensor_data], axis=1)

readings_df
```

```python
readings_df['country Code'] = country_code
```

```python
Courses_Statistics = readings_df.groupby('country Code').agg({'Temperture': ['mean', 'max', 'min']})
Courses_Statistics.columns = ['average_temp', 'max_temp', 'min_temp']
Courses_Statistics
```
